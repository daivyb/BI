```python
import pandas as pd
df = pd.read_csv("Real Estate Dataset.csv", sep=";")
```

```python
#"Int64" → especial de pandas, permite NaN

# provision we need to change it to objetct values
df["provision"] = df["provision"].astype("object")

# Para poder visualizar correctamente los datos, es necesario cambiar el tipo de dato de algunas columnas. En este caso, las columnas "year_built", "last_reconstruction", "total_floors", "floor", "balkonies", "loggia" y "rooms" se convertirán a tipos de datos enteros int64 o Int64, dependiendo de si permiten valores NaN o no
df["year_built"] = df["year_built"].astype("Int64")
df["last_reconstruction"] = df["last_reconstruction"].astype("Int64")
df["total_floors"] = df["total_floors"].astype("Int64")
df["floor"] = df["floor"].astype("Int64")
df["balkonies"] = df["balkonies"].astype("Int64")
df["loggia"] = df["loggia"].astype("Int64")
df["rooms"] = df["rooms"].astype("int64")

# Cambiar coma decimal a punto decimal y convertir a tipo de dato float.
# Las columnas index	environment	quality_of_living	safety	transport	services	relax	
df["area"] = df["area"].str.replace(",", ".").astype(float)
df["index"] = df["index"].str.replace(",", ".").astype(float)
df["environment"] = df["environment"].str.replace(",", ".").astype(float)
df["quality_of_living"] = df["quality_of_living"].str.replace(",", ".").astype(float)
df["safety"] = df["safety"].str.replace(",", ".").astype(float)
df["transport"] = df["transport"].str.replace(",", ".").astype(float)
df["services"] = df["services"].str.replace(",", ".").astype(float)
df["relax"] = df["relax"].str.replace(",", ".").astype(float)

df["index"].unique()
```

Limpieza de datos

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df['price'])
plt.title('Diagrama de cajas de Price')
plt.ylabel('Price')
plt.show()
```

```python
# Calcular Q1, Q3 e IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Crear columna 'is_outlier' (True si es outlier, False si no)
df['price_outlier'] = (df['price'] < lower_bound) | (df['price'] > upper_bound)
df['price_outlier'] = df['price_outlier']

# Revisar las primeras filas
print(df[['price', 'price_outlier']].head())
```

```python
# Validaciones adicionales
df["area"].value_counts(dropna=False)
df['total_floors'].sort_values(ascending=True)
df["last_reconstruction"].describe()
df[df["price"] == 9936000][["price", "area"]]
df["area"].sort_values(ascending=True).head(300)


df["cellar"].value_counts(dropna=False)

```

```python
# EDA Inicial

# Reemplazar valores irreales en la columna area por 0
df['area'] = df['area'].apply(lambda x: x if 35 <= x < 900 else 0)

# Reemplazar valores negativos en la columna "floor" por 0
df['floor'] = df['floor'].apply(lambda x: x if x >= 0 else 0)

# Reemplazar valores irreales en la fecha por 0
# Esto ayuda a limpiar los datos y evitar que valores incorrectos afecten el análisis posterior

df['year_built'] = df['year_built'].apply(lambda x: x if 1900 <= x <= 2026 else 0)

# Crear columnas de indicadores de valores faltantes para las columnas
# Ayuda a identificar rápidamente qué filas tienen valores faltantes en estas columnas específicas y facilita el análisis de la calidad de los datos.  

cols = ["loggia", "balkonies", "last_reconstruction", "orientation", "construction_type", "certificate", "condition", "area", "index", "environment", "quality_of_living", "safety", "transport", "services", "relax", "energy_costs"]

for col in cols:
    df[col + "_missing"] = df[col].isna()


# Reemplazar valores faltantes (NaN) por "Unknown"

cols_to_fill = ["orientation", "construction_type", "certificate", "condition"]

for col in cols_to_fill:
    df[col] = df[col].fillna("Unknown")

# Reemplazar valores faltantes (NaN) por 0 en la columna "year_built" y convertir a tipo de dato int64

cols_to_fill_zero = ["year_built", "last_reconstruction", "total_floors", "floor", "balkonies", "loggia", "area", "index", "environment", "quality_of_living", "safety", "transport", "services", "relax", "energy_costs"]

for col in cols_to_fill_zero:
    df[col] = df[col].fillna(0)
    df[col] = df[col].astype("int64")
```
